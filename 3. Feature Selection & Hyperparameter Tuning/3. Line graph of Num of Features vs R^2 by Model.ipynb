```python
import pandas as pd
import matplotlib.pyplot as plt

# Read in the data from the CSV files
lasso = pd.read_csv('results_lasso.csv')
ridge = pd.read_csv('results_ridge.csv')
elasticnet = pd.read_csv('results_elasticnet.csv')
mlr = pd.read_csv('results_mlr.csv')
dt = pd.read_csv('results_dt.csv')
xgboost = pd.read_csv('results_xgboost.csv')
rf = pd.read_csv('results_rf.csv')
lgbm = pd.read_csv('results_lgbm.csv')
kernelridge = pd.read_csv('results_kernel_ridge.csv')
svr = pd.read_csv('results_svr.csv')
mlp = pd.read_csv('results_mlp.csv')
stacked_ensemble = pd.read_csv('results_stacked_ensemble.csv')  # New dataframe

# Create a list of the models and dataframes
models = ['XGBoost','Stacked Ensemble','LightGBM','Decision Tree','Random Forest','ElasticNet', 'Lasso', 'MLR',  'Ridge','Kernel Ridge', 'SVR', 'MLP']  # Updated
dataframes = [xgboost,stacked_ensemble, lgbm, dt,rf,elasticnet,lasso, mlr, ridge, kernelridge, svr, mlp]  # Updated

colors = ['red', 'blue', 'green', 'orange', 'purple', 'pink', 'brown', 'gray', 'olive', 'teal', 'navy', 'maroon']

# Set up the figure
plt.figure(figsize=(8, 6))
plt.subplots_adjust(left=0.1, right=0.9, bottom=0.1, top=0.9)

for i, df in enumerate(dataframes):
    plt.plot(df['importance_rank'], df['test_r2'], label=models[i], color=colors[i])

# Set the axis labels and title
plt.xlabel('Number of Features', fontsize=16)
plt.ylabel('R2 Score', fontsize=16)
plt.title('Number of Features vs R^2 Score by Model', fontsize=16)

# Add grid lines to the plot
plt.grid(True)

# Set the legend title and properties
legend = plt.legend(loc='center left', bbox_to_anchor=(1, 0.40), frameon=True, framealpha=0.8, fontsize=16, title='Model Performance Ranking')
legend.get_title().set_fontsize(16)

plt.xticks(range(0, max(df['importance_rank'])+1, 1), fontsize=14)
plt.yticks(fontsize=14)

# Set the y-axis limits
plt.xlim(1, 10)
plt.ylim(0.45, 1)

# Show the plot
plt.show()
```
